## Progressbar

```python
def log_progress(sequence, every=None, size=None, name='Items'):
    from ipywidgets import IntProgress, HTML, VBox
    from IPython.display import display

    is_iterator = False
    if size is None:
        try:
            size = len(sequence)
        except TypeError:
            is_iterator = True
    if size is not None:
        if every is None:
            if size <= 200:
                every = 1
            else:
                every = int(size / 200)     # every 0.5%
    else:
        assert every is not None, 'sequence is iterator, set every'

    if is_iterator:
        progress = IntProgress(min=0, max=1, value=1)
        progress.bar_style = 'info'
    else:
        progress = IntProgress(min=0, max=size, value=0)
    label = HTML()
    box = VBox(children=[label, progress])
    display(box)

    index = 0
    try:
        for index, record in enumerate(sequence, 1):
            if index == 1 or index % every == 0:
                if is_iterator:
                    label.value = '{name}: {index} / ?'.format(
                        name=name,
                        index=index
                    )
                else:
                    progress.value = index
                    label.value = u'{name}: {index} / {size}'.format(
                        name=name,
                        index=index,
                        size=size
                    )
            yield record
    except:
        progress.bar_style = 'danger'
        raise
    else:
        progress.bar_style = 'success'
        progress.value = index
        label.value = "{name}: {index}".format(
            name=name,
            index=str(index or '?')
        )
```

```python
class MRIDataset(DataLoader):
    def __init__(self, data_list, acceleration, center_fraction, use_seed):
        self.data_list = data_list
        self.acceleration = acceleration
        self.center_fraction = center_fraction
        self.use_seed = use_seed

    def __len__(self):
        return len(self.data_list)

    def __getitem__(self, idx):
        subject_id = self.data_list[idx]
        return get_epoch_batch(subject_id, self.acceleration, self.center_fraction, self.use_seed)
```

```python
def get_epoch_batch(subject_id, acc, center_fract, use_seed=True):
    ''' random select a few slices (batch_size) from each volume'''

    fname, rawdata_name, slice = subject_id  
    
    with h5py.File(rawdata_name, 'r') as data:
        rawdata = data['kspace'][slice]
                      
    slice_kspace = T.to_tensor(rawdata).unsqueeze(0)
    S, Ny, Nx, ps = slice_kspace.shape

    # apply random mask
    shape = np.array(slice_kspace.shape)
    mask_func = MaskFunc(center_fractions=[center_fract], accelerations=[acc])
    seed = None if not use_seed else tuple(map(ord, fname))
    mask = mask_func(shape, seed)
      
    # undersample
    masked_kspace = torch.where(mask == 0, torch.Tensor([0]), slice_kspace)
    masks = mask.repeat(S, Ny, 1, ps)

    img_gt, img_und = T.ifft2(slice_kspace), T.ifft2(masked_kspace)
    
    
    # perform data normalization which is important for network to learn useful features
    # during inference there is no ground truth image so use the zero-filled recon to normalize
    norm = T.complex_abs(img_und).max()
    if norm < 1e-6: norm = 1e-6
    
    # normalized data
    img_gt, img_und, rawdata_und = img_gt/norm, img_und/norm, masked_kspace/norm
    
    #crop image?
    img_gt = T.complex_center_crop(img_gt, [320, 320])
    img_und = T.complex_center_crop(img_und, [320, 320])
    rawdata_und = T.complex_center_crop(rawdata_und, [320, 320])
    masks = T.complex_center_crop(masks, [320, 320])
        
    return img_gt.squeeze(0), img_und.squeeze(0), masks.squeeze(0), norm
```

```python
def load_data_path(train_data_path, val_data_path):
    """ Go through each subset (training, validation) and list all 
    the file names, the file paths and the slices of subjects in the training and validation sets 
    """

    data_list = {}
    train_and_val = ['train', 'val']
    data_path = [train_data_path, val_data_path]
      
    for i in range(len(data_path)):

        data_list[train_and_val[i]] = []
        
        which_data_path = data_path[i]
    
        for fname in sorted(os.listdir(which_data_path)):
            
            subject_data_path = os.path.join(which_data_path, fname)
                     
            if not os.path.isfile(subject_data_path): continue 
            
            with h5py.File(subject_data_path, 'r') as data:
                num_slice = data['kspace'].shape[0]
                
            # the first 5 slices are mostly noise so it is better to exlude them
            data_list[train_and_val[i]] += [(fname, subject_data_path, slice) for slice in range(5, num_slice)]
    
    return data_list    
```

```python
def load_data_path2(data_path):
    """ Go through each subset (training, validation) and list all 
    the file names, the file paths and the slices of subjects in the training and validation sets 
    """

    data_list = {}
    train_and_val = ['train', 'val']
    list_dir = sorted(os.listdir(data_path))
    x = int(len(list_dir)/10*8) - 1
    list_dir_split = [list_dir[0: x], list_dir[x: -1]]
      
    for i in range(len(list_dir_split)):

        data_list[train_and_val[i]] = []
        
        which_list_dir = list_dir_split[i]
    
        for fname in which_list_dir:
            
            subject_data_path = os.path.join(data_path, fname)
                     
            if not os.path.isfile(subject_data_path): continue 
            
            with h5py.File(subject_data_path, 'r') as data:
                num_slice = data['kspace'].shape[0]
                
            # the first 5 slices are mostly noise so it is better to exlude them
            data_list[train_and_val[i]] += [(fname, subject_data_path, slice) for slice in range(5, num_slice)]
    
    return data_list 
```

## Model

```python
class ConvBlock(nn.Module):
    """
    A Convolutional Block that consists of two convolution layers each followed by
    instance normalization, relu activation and dropout.
    """

    def __init__(self, in_chans, out_chans, drop_prob):
        """
        Args:
            in_chans (int): Number of channels in the input.
            out_chans (int): Number of channels in the output.
            drop_prob (float): Dropout probability.
        """
        super().__init__()

        self.in_chans = in_chans
        self.out_chans = out_chans
        self.drop_prob = drop_prob

        self.layers = nn.Sequential(
            nn.Conv2d(in_chans, out_chans, kernel_size=3, padding=1),
            nn.InstanceNorm2d(out_chans),
            nn.ReLU(),
            nn.Dropout2d(drop_prob),
            nn.Conv2d(out_chans, out_chans, kernel_size=3, padding=1),
            nn.InstanceNorm2d(out_chans),
            nn.ReLU(),
            nn.Dropout2d(drop_prob)
        )

    def forward(self, input):
        """
        Args:
            input (torch.Tensor): Input tensor of shape [batch_size, self.in_chans, height, width]
        Returns:
            (torch.Tensor): Output tensor of shape [batch_size, self.out_chans, height, width]
        """
        return self.layers(input)

    def __repr__(self):
        return f'ConvBlock(in_chans={self.in_chans}, out_chans={self.out_chans}, ' \
            f'drop_prob={self.drop_prob})'
```

```python
class UnetModel(nn.Module):
    """
    PyTorch implementation of a U-Net model.
    This is based on:
        Olaf Ronneberger, Philipp Fischer, and Thomas Brox. U-net: Convolutional networks
        for biomedical image segmentation. In International Conference on Medical image
        computing and computer-assisted intervention, pages 234–241. Springer, 2015.
    """

    def __init__(self, in_chans, out_chans, chans, num_pool_layers, drop_prob):
        """
        Args:
            in_chans (int): Number of channels in the input to the U-Net model.
            out_chans (int): Number of channels in the output to the U-Net model.
            chans (int): Number of output channels of the first convolution layer.
            num_pool_layers (int): Number of down-sampling and up-sampling layers.
            drop_prob (float): Dropout probability.
        """
        super().__init__()

        self.in_chans = in_chans
        self.out_chans = out_chans
        self.chans = chans
        self.num_pool_layers = num_pool_layers
        self.drop_prob = drop_prob

        self.down_sample_layers = nn.ModuleList([ConvBlock(in_chans, chans, drop_prob)])
        ch = chans
        for i in range(num_pool_layers - 1):
            self.down_sample_layers += [ConvBlock(ch, ch * 2, drop_prob)]
            ch *= 2
        self.conv = ConvBlock(ch, ch, drop_prob)

        self.up_sample_layers = nn.ModuleList()
        for i in range(num_pool_layers - 1):
            self.up_sample_layers += [ConvBlock(ch * 2, ch // 2, drop_prob)]
            ch //= 2
        self.up_sample_layers += [ConvBlock(ch * 2, ch, drop_prob)]
        self.conv2 = nn.Sequential(
            nn.Conv2d(ch, ch // 2, kernel_size=1),
            nn.Conv2d(ch // 2, out_chans, kernel_size=1),
            nn.Conv2d(out_chans, out_chans, kernel_size=1),
        )

    def forward(self, input):
        """
        Args:
            input (torch.Tensor): Input tensor of shape [batch_size, self.in_chans, height, width]
        Returns:
            (torch.Tensor): Output tensor of shape [batch_size, self.out_chans, height, width]
        """
        stack = []
        output = input
        # Apply down-sampling layers
        for layer in self.down_sample_layers:
            output = layer(output)
            stack.append(output)
            output = F.max_pool2d(output, kernel_size=2)

        output = self.conv(output)

        # Apply up-sampling layers
        for layer in self.up_sample_layers:
            output = F.interpolate(output, scale_factor=2, mode='bilinear', align_corners=False)
            output = torch.cat([output, stack.pop()], dim=1)
            output = layer(output)
        return self.conv2(output)
```

```python
class Model3(nn.Module):
    def __init__(self):
        """
        
        We have 12 layers.
        
        
        self.layer1 = nn.Sequential(
            nn.Conv2d(1, 4, kernel_size=5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(4) 
            )
            
        When we look at this layer, nn.Sequential() do each step in sequence. So it does one after the other.
        Here, it is conv2d, then relu, then BatchNorm.#
        
        The argument in Conv2d is the following: 
        (number of input channels, number of output channels, kernal_size, stridge, and padding)
        
        Basically, we are using a channel of 1 since it is greyscaled image.
        
        If we used K-complex values, the channels will be 2. (imaginary, and real values I believe)
        
        If we used colored images, the number of channels will be 3. Red, Green, and Blue Channels.
        
        A channel is a "feature" of the image. The colored images are defined by RGB values, so we have 3 channels.
        
        We are using greyscale, so we only have one channel.
        
        Now, let's get to the actual convolutional part.
        
        The kernel_size is the size of the filter that we use.
        
        If the kernel_size is 5, we are using 5 x 5 filter.
        
        Since our first layer, the number of input channels is 1 and the output channels is 4,
        we are applying 4 different "filters" to create our 4 output channels.
        
        Stride refers to how many pixels we move after we apply our filters.
        
        Padding is used to create edges in the image. I've added padding of 2 to keep the original size.
        
        Everybody will know what a relu is.
        
        Batchnorm normalizes the image. I don't know really what it does but I've added after every image
        
        Our second layer has input channel of 4 and output channel of 16. So it creates 4 filters and applies it four input channels.
        
        We expand our channels to 256 and then we decrease the channels until we get to channel of 1, our output layer.
        
        
        

        """
        super(Model3,self).__init__()
                
        self.layer1 = nn.Sequential(
            nn.Conv2d(1, 4, kernel_size=5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(4) 
            )
    
        
        self.layer2 = nn.Sequential(
            nn.Conv2d(4, 16, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(16) 
            )
     
    
        self.layer3 = nn.Sequential(
            nn.Conv2d(16, 32, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(32) 
            )
       
        
        self.layer4 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(64)
            )
        
        self.layer5= nn.Sequential(
            nn.Conv2d(64, 128, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(128)
            )
        
        self.layer6= nn.Sequential(
            nn.Conv2d(128, 256, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(256) 
            )
        
        self.layer7= nn.Sequential(
            nn.Conv2d(256, 128, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(128)
            )
        
        self.layer8= nn.Sequential(
            nn.Conv2d(128, 64, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(64)  
            )
        

        self.layer9= nn.Sequential(
            nn.Conv2d(64, 32, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(32) 
            )
     

        self.layer10= nn.Sequential(
            nn.Conv2d(32, 16, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(16) 
            )
            
        
        self.layer11= nn.Sequential(
            nn.Conv2d(16, 4, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(4) 
            )
        
        self.layer12= nn.Sequential(
            nn.Conv2d(4, 1, kernel_size = 5, stride=1, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(1)
        )
            
    
        
    def forward(self, x):

        out = self.layer1(x)
        out = self.layer2(out)

        out = self.layer3(out)
        out = self.layer4(out)
        out = self.layer5(out)
        out = self.layer6(out)
        out = self.layer7(out)
        out = self.layer8(out)
        out = self.layer9(out)
        out = self.layer10(out)
        out = self.layer11(out)
        out = self.layer12(out)

        return out
```

## Generic code to train the data

```python
def generic_code(model, optimiser):

    # Define a function inside another function
    def train_step(x_batch, y_batch):
        optimiser.zero_grad()
        y_pred = model(x_batch)            # forward pass
        loss = F.l1_loss(y_batch, y_pred)  # calculate loss value
        #loss = ssim(y_batch.reshape(1,320,320).detach().cpu().numpy(), y_pred.reshape(1,320,320).detach().cpu().numpy())  
        loss.backward()                    # autograd
        optimiser.step()                   # update parameters  
        return loss.item()                 # return the loss

    # Return the newly defined function
    return train_step                      # Return a function
```

## SSIM

```python
from skimage.measure import compare_ssim 
def ssim(gt, pred):
    """ Compute Structural Similarity Index Metric (SSIM). """
    return compare_ssim(
        gt.transpose(1, 2, 0), pred.transpose(1, 2, 0), multichannel=True, data_range=gt.max()
    )
```

## Save & Load Model

```python
def save_model(PATH, model, optimiser, acc, epoch, losses, val_losses, val_ssim):
    torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimiser_state_dict': optimiser.state_dict(),
            'acceleration': acc,
            'losses': losses,
            'val_losses': val_losses,
            'val_ssim': val_ssim
            }, "{}model_{}af.pt".format(PATH, acc))
```

```python
def load_model(PATH, model, optimiser=None):
    checkpoint = torch.load(PATH)
 
    model.load_state_dict(checkpoint['model_state_dict'])
    if optimiser:
        optimiser.load_state_dict(checkpoint['optimiser_state_dict'])
        acc = checkpoint['acceleration']
        epoch = checkpoint['epoch']
        losses = checkpoint['losses']
        val_losses = checkpoint['val_losses']
        val_ssim = checkpoint['val_ssim']
    
        return model, optimiser, acc, epoch, losses, val_losses, val_ssim
    else:
        return model
```

## Load Data

```python
def load_data(data_path, acc):
    data_list = load_data_path2(data_path)  # first load all file names, paths and slices.

    cen_fract = 0.04
    seed = False  # random masks for each slice 
    num_workers = 12  # data loading is faster using a bigger number for num_workers. 0 means using one cpu to load data
  
    # Create data loader for training set. It applies same to validation set as well
    train_dataset = MRIDataset(data_list['train'], acceleration=acc, center_fraction=cen_fract, use_seed=seed)
    train_loader = DataLoader(train_dataset, shuffle=True, batch_size=5, num_workers=num_workers) 

    # Create data loader for validation set
    val_dataset = MRIDataset(data_list['val'], acceleration=acc, center_fraction=cen_fract, use_seed=seed)
    val_loader = DataLoader(val_dataset, shuffle=True, batch_size=5, num_workers=num_workers)

    # Create data loader for calculating ssim before train
    val_dataset = MRIDataset(data_list['val'], acceleration=acc, center_fraction=cen_fract, use_seed=seed)
    val_loader = DataLoader(val_dataset, shuffle=True, batch_size=5, num_workers=num_workers)
    
    return train_loader, val_loader
```

## Train functions

```python
def train_model(data_path, device, model, optimiser, acc, lr, epoches, epoch_cont, losses, val_losses, val_ssim):
    
    train_loader, val_loader = load_data(data_path, acc)
    
    train_step = generic_code(model, optimiser)
    orig_ssim = list()
    last_val_ssim = list()

    starting_time = time.time()
    for epoch in log_progress(range(epoch_cont, epoches), every=1, name="Epoch"):
        model.train()
        
        for iteration, sample in enumerate(log_progress(train_loader,
                                                        every=1,
                                                        name="{} - Train Data".format(epoch+1))):
            #print("epoch:", epoch, "- iteration:", iteration)
            
            img_gt, img_und, masks, norm = sample
           
            img_und = T.complex_abs(img_und).squeeze()
            img_gt = T.complex_abs(img_gt).squeeze()

            img_gt = img_gt.unsqueeze(1).to(device)
            img_und= img_und.unsqueeze(1).to(device)

            # List to store the loss for every epoch
            loss = train_step(img_und, img_gt)
            if iteration % 10 == 0:
                losses.append(loss)

        last_val_ssim.clear()  # Clear the list, only store ssim value for last epoch
        for iteration_val, sample_val in enumerate(log_progress(val_loader,
                                                                every=1,
                                                                name="{} - Validate Data".format(epoch+1))):
                #print(iteration_val)
                
                img_gt_val, img_und_val, masks_val, norm_val = sample_val
               
                img_gt_val = T.complex_abs(img_gt_val).squeeze()
                img_und_val = T.complex_abs(img_und_val).squeeze()

                img_gt_val = img_gt_val.unsqueeze(1).to(device)
                img_und_val = img_und_val.unsqueeze(1).to(device)
                
                img_gt_pred = model(img_und_val)
                val_loss = F.l1_loss(img_gt_val, img_gt_pred)
                
                val_loss = val_loss.item()
                if iteration_val % 10 == 0:
                    # List to store the loss for every epoch
                    val_losses.append(val_loss)
                  
                # List to store the ssim value before training the data
                orig_ssim.append(ssim(img_gt_val.squeeze().cpu().numpy(),
                                     img_und_val.squeeze().cpu().detach().numpy()))
                if iteration_val % 10 == 0:
                    # List to store the ssim value after training 
                    val_ssim.append(ssim(img_gt_val.squeeze().cpu().numpy(),
                                     img_gt_pred.squeeze().cpu().detach().numpy()))
                # List to store the ssim value for last epoch
                last_val_ssim.append(ssim(img_gt_val.squeeze().cpu().numpy(),
                                     img_gt_pred.squeeze().cpu().detach().numpy()))
    
        
        # Save model:
        dir_path = ""
        save_model(dir_path, model, optimiser, acc, epoch, losses, val_losses, val_ssim)

    print('Training done')
    print('Training time taken: ', time.time() - start_time)
    
    return model, losses, val_losses, val_ssim, orig_ssim, last_val_ssim
```

```python
def test_reconstruction(file_path, model_path, acc):
    """
    Function to create reconstructions for one model (acc 4 or 8)
    and show 5 slices of the reconstructed data

    The function does not save anything
    """

    # Load model:
    device = 'cuda:0' if torch.cuda.is_available() else 'cpu'

    in_chans = 1 
    out_chans = 1
    chans = 32
    num_pool_layers = 4
    drop_prob = 0.0
    model = UnetModel(in_chans, out_chans, chans, num_pool_layers, drop_prob).to(device)

    model = load_model(model_path, model)
    
    model.to(device)
    model.eval()
    
    # Load test data:
    for fname in sorted(os.listdir(file_path)):
        subject_path = os.path.join(file_path, fname)
        with h5py.File(subject_path,  "r") as hf:
            if acc == 4:
                volume_kspace = hf['kspace_4af'][()]
            elif acc == 8:
                volume_kspace = hf['kspace_8af'][()]
            
            volume_kspace2 = T.to_tensor(volume_kspace)     # Convert from numpy array to pytorch tensor
            volume_image = T.ifft2(volume_kspace2)          # Apply Inverse Fourier Transform to get the complex image
            volume_image_abs = T.complex_abs(volume_image)  # Compute absolute value to get a real image
            
            volume_image_output = np.zeros(shape=(len(volume_image_abs),320,320))
            
            for i, slice in enumerate(log_progress(volume_image_abs,
                                                   every=1,
                                                   name="Reconstructing {}".format(fname))):
                
                slice = T.center_crop(slice, [320, 320]).reshape(1,1,320,320).to(device)
                img_gt_pred = model(slice)
                volume_image_output[i] = img_gt_pred.cpu().detach().numpy()
            
            # Show original and reconstriction (5 slices)
            print("Original")
            show_slices(T.center_crop(volume_image_abs, [320, 320]) , [5, 10, 20, 30], cmap='gray')
            
            print("Reconstructed")
            show_slices(volume_image_output, [5, 10, 20, 30], cmap='gray')
```

## Reconstruct

```python
def save_reconstructions(reconstructions, out_dir):
    """
    Saves the reconstructions from a model into h5 files that is appropriate for submission
    to the leaderboard.
    Args:
        reconstructions (dict[str, np.array]): A dictionary mapping input filenames to
            corresponding reconstructions (of shape num_slices x height x width).
        out_dir (pathlib.Path): Path to the output directory where the reconstructions
            should be saved.
    """
    for fname, recons in reconstructions.items():
        subject_path = os.path.join(out_dir, fname)
        print(subject_path)
        with h5py.File(subject_path, 'w') as f:
            f.create_dataset('reconstruction', data=recons)
```

```python
def reconstruct_images(file_path, model_path_4af, model_path_8af, out_dir):
    """
    Loads two trained models, one with acceleration of 4 and one with acceleration of 8.
    Loads the h5 test files, reconstructs each slice for kspace_4af and kspace_8af, and
    saves a dictionary object containing the reconstructions to a h5 file with the same name.

    Output h5-file structure:
        h5-file = {
            'recon_4af' : volume_image_output - (slices, 350, 350)
            'recon_8af' : volume_image_output - (slices, 350, 350)
        }
    """
    
    # Load model:
    device = 'cuda:0' if torch.cuda.is_available() else 'cpu'

    in_chans = 1 
    out_chans = 1
    chans = 32
    num_pool_layers = 4
    drop_prob = 0.0
    model = UnetModel(in_chans, out_chans, chans, num_pool_layers, drop_prob).to(device)
    
    model_4af = load_model(model_path_4af, model)
    model_8af = load_model(model_path_8af, model)
    
    model_4af.to(device)
    model_8af.to(device)
    model_4af.eval()
    model_8af.eval()
    
    # Load test data:
    for fname in sorted(os.listdir(file_path)):
        subject_path = os.path.join(file_path, fname)
        
        with h5py.File(subject_path,  "r") as hf:
            final_output = {}
    
            # Reconstruct data:
            for x in ['kspace_4af', 'kspace_8af']:
                volume_kspace = hf[x][()]                       # Get volume kspace
                volume_kspace2 = T.to_tensor(volume_kspace)     # Convert from numpy array to pytorch tensor
                volume_image = T.ifft2(volume_kspace2)          # Apply Inverse Fourier Transform to get the complex image
                volume_image_abs = T.complex_abs(volume_image)  # Compute absolute value to get a real image

                volume_image_output = np.zeros(shape=(len(volume_image_abs),320,320))  # Create empty dict

                for i, slice in enumerate(log_progress(volume_image_abs,
                                                       every=1,
                                                       name="Reconstructing {}".format(fname))):
                  
                    slice = T.center_crop(slice, [320, 320]).reshape(1,1,320,320).to(device)
                    
                    # Use correct model
                    if x == 'kspace_4af':
                        img_gt_pred = model_4af(slice)
                    elif x == 'kspace_8af':
                        img_gt_pred = model_8af(slice)
                        
                    volume_image_output[i] = img_gt_pred.cpu().detach().numpy()
                
                # Put under correct key in dictionary
                if x == 'kspace_4af':
                    final_output["recon_4af"] = volume_image_output
                elif x == 'kspace_8af':
                    final_output["recon_8af"] = volume_image_output

                # Show original and reconstriction (5 slices)
                print("Original - {}".format(x))
                show_slices(T.center_crop(volume_image_abs, [320, 320]) , [5, 10, 20, 30], cmap='gray')

                print("Reconstructed - {}".format(x))
                show_slices(volume_image_output, [5, 10, 20, 30], cmap='gray')
            
            # Save reconstruction
            if not (os.path.exists(out_dir)): os.makedirs(out_dir)
            save_reconstructions({fname: final_output}, out_dir)
```
